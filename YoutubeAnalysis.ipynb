```python
def categorize_col(df, col, labels):
    # If already categorized, skip processing
    if df[col].dtype == 'object' and df[col].str.contains(':').any():
        # Clean whitespace
        df[col] = df[col].str.strip()
        
        # Convert time strings like "13:45" to hour integers
        df[col] = pd.to_datetime(df[col], format='%H:%M').dt.hour
        
        # Define bin edges
        edges = [0, 12, 16, 20, 24]
        
        # Categorize hours into time of day
        df[col] = pd.cut(df[col], bins=edges, labels=labels, right=False, duplicates='drop')
    
    return df
```

```python
# Define your labels
labels = ['Noon', 'Afternoon', 'Evening', 'Night']

# Example usage
df = categorize_col(df, 'Time', labels)
print(df['Time'].unique())
```

```python


# Step 1: No filtering — include all views starting from 0
max_views = df["views"].max()

# Step 2: Define bins from 0 to max views in steps of 10M
bins = list(range(0, int(max_views) + 10_000_000, 10_000_000))

# Step 3: Plot histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x="views", bins=bins, color="#6495ED", kde=False)

# Step 4: Customize plot
plt.title("Histogram of Video Views (0M+ Range)")
plt.xlabel("Views")
plt.ylabel("Number of Videos")

# Step 5: Format xticks as '0M–10M', '10M–20M', ...
plt.xticks(
    bins, 
    [f"{i//1_000_000}M–{(i+10_000_000)//1_000_000}M" for i in bins], 
    rotation=45
)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Filter data above 10 million views
filtered_df = df[df["views"] >= 10_000_000]

# Step 2: Define bins (you can go up as high as you want)
bins = list(range(10_000_000, filtered_df["views"].max() + 10_000_000, 10_000_000))

# Step 3: Plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=filtered_df, x="views", bins=bins, color="#6495ED", kde=False)

# Step 4: Customize plot
plt.title("Histogram of Video Views (10M+ Range)")
plt.xlabel("Views")
plt.ylabel("Number of Videos")
plt.xticks(bins, [f"{i//1_000_000}M-{(i+10_000_000)//1_000_000}M" for i in bins], rotation=45)
plt.tight_layout()
plt.show()
```
